Send a request to CityBikes for the city of your choice.

```python
# API request for bike station data for the city for Abu Dhabi
URL = "http://api.citybik.es/v2/networks/adcb-bikeshare"

params_dict = { 'fields' : 'id,name,stations,location' }

city_bikes_networks = requests.get(URL , params=params_dict).json()
city_bikes_networks 
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python

import pandas as pd


AbuDhabi_bikes = []

for station in city_bikes_networks['network']['stations']:
        AbuDhabi_bikes.append([station['empty_slots'],station['free_bikes'],station['latitude'], station['longitude']])
```

Put your parsed results into a DataFrame.

```python
df = pd.DataFrame(AbuDhabi_bikes, columns= ['empty_slots', 'free_bikes', 'latitude', 'longitude'])
df
```

```python
df.to_csv('bike_locationsfull.csv', index=False)
```
